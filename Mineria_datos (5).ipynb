# IMPUTACIÓN DE VALORES

Valores floats se eliminan con la media

```python
import pandas as pd
import gdown

# Descargar el archivo desde Google Drive
file_id = "1bYlcwgglmefBA3vmI36J9_VcRWOk1NQq"
url = f'https://drive.google.com/uc?id={file_id}'
output = 'weatherAUS.xlsx'
gdown.download(url, output, quiet=False)

# Leer el archivo Excel en un DataFrame
df = pd.read_excel("weatherAUS.xlsx")

# Imputar valores faltantes en variables numéricas con la media
columnas_numericas = ['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine', 'WindGustSpeed',
                      'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm', 'Pressure9am',
                      'Pressure3pm', 'Cloud9am', 'Cloud3pm', 'Temp9am', 'Temp3pm', 'RISK_MM']

df[columnas_numericas] = df[columnas_numericas].fillna(df[columnas_numericas].mean())

# Ahora, df contiene el dataframe con los valores faltantes imputados utilizando la media
```

Los tipos object se hacen con la moda

```python
# Importar pandas
import pandas as pd

# Leer el archivo Excel en un DataFrame
df = pd.read_excel("weatherAUS.xlsx")

# Identificar las columnas de tipo 'object'
columnas_object = df.select_dtypes(include='object').columns

# Imputar valores faltantes en las columnas de tipo 'object' con la moda
df[columnas_object] = df[columnas_object].fillna(df[columnas_object].mode().iloc[0])

# Ahora, df contiene el dataframe con los valores faltantes imputados en las columnas 'object' utilizando la moda
```

Datos con muchos nulos con KNN

```python
# Importar pandas y SimpleImputer de sklearn
import pandas as pd
from sklearn.impute import KNNImputer

# Leer el archivo Excel en un DataFrame
df = pd.read_excel("weatherAUS.xlsx")

# Definir las columnas con muchos nulos
columnas_con_nulos = ['Evaporation', 'Sunshine', 'Cloud9am', 'Cloud3pm']

# Crear un imputador KNN
imputer = KNNImputer()

# Imputar los valores faltantes en las columnas seleccionadas
df_imputado = df.copy()  # Realiza una copia del dataframe para preservar el original
df_imputado[columnas_con_nulos] = imputer.fit_transform(df[columnas_con_nulos])

# Ahora, df_imputado contiene el dataframe con los valores faltantes imputados utilizando KNN
```

```python
# Imprimir un resumen estadístico del DataFrame después de la imputación con KNN
print(df_imputado.describe())
```

# Análisis de los valores imputados a través de los tres métodos según su tipo


**Count**: Muestra el número de valores no nulos en cada columna. Después de la imputación con KNN, todas las columnas tienen el mismo número de filas, lo que indica que no hay valores faltantes en ninguna de las columnas.

**Mean**: Indica la media de cada columna. La media se ha calculado para cada columna después de imputar los valores faltantes con KNN. Observamos que las medias son similares a las de antes de la imputación, lo que sugiere que KNN ha imputado los valores de manera efectiva sin afectar significativamente las estadísticas de las columnas.

**Std** (Desviación estándar): Muestra la dispersión de los datos alrededor de la media. Al igual que con la media, las desviaciones estándar son similares antes y después de la imputación, lo que sugiere que la variabilidad de los datos se ha mantenido después de la imputación.

**Min y Max**: Indican los valores mínimo y máximo de cada columna. Estos valores no han cambiado después de la imputación con KNN, lo que indica que no se han introducido valores extremos o atípicos durante el proceso de imputación.

# **Identificar outliers y valores faltantes en los datos**

Utilizaremos gráficos de caja (box plots)  y diagramas de dispersión (scatter plots) para visualizar las distribuciones de datos y identificar puntos fuera de lo común.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Supongamos que df es nuestro DataFrame

# Paso 1: Calcular el rango intercuartílico (IQR)
Q1 = df.quantile(0.25, numeric_only=True)
Q3 = df.quantile(0.75, numeric_only=True)
IQR = Q3 - Q1

# Paso 2: Identificar valores atípicos
outliers = ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR)))

# Paso 3: Visualizar distribuciones de datos
# Gráficos de caja (box plots) para cada columna numérica
plt.figure(figsize=(12, 6))
sns.boxplot(data=df)
plt.title('Box Plot de las Variables Numéricas')
plt.xticks(rotation=45)
plt.show()

# Visualizar los valores atípicos identificados
print("Valores Atípicos:")
print(outliers.sum())
```

### **Análisis de valores atípicos utilizando el método del rango intercuartílico (IQR)**

Cada columna del DataFrame tiene un recuento de valores atípicos identificados.
Las columnas *'Cloud3pm'* y '*Cloud9am'* no tienen valores atípicos identificados, ya que su recuento es 0.
Otras columnas como *'Evaporation', 'Humidity3pm', 'Humidity9am', 'MaxTemp', 'MinTemp', 'Pressure3pm', 'Pressure9am', 'RISK_MM', 'Rainfall', 'Temp3pm', 'Temp9am', 'WindGustSpeed', 'WindSpeed3pm', y 'WindSpeed9am'* tienen valores atípicos identificados.

Puede ser necesario tomar medidas adicionales, como la imputación de valores, la eliminación de valores atípicos o la consideración de métodos estadísticos más avanzados para manejarlos adecuadamente.

# Recomendaciones de tratamiento de Outliers y Missing Values:


**Eliminación de filas o columnas con valores faltantes:** Si los valores faltantes representan una pequeña fracción de tus datos y eliminarlos no afecta significativamente tu análisis, puedes optar por eliminar las filas o columnas que contienen valores faltantes.

**Imputación con estadísticas descriptivas:** Puedes imputar los valores faltantes utilizando estadísticas descriptivas como la media, la mediana o la moda de la columna respectiva. Esto puede ser útil para variables numéricas o categóricas.

**Imputación con modelos de aprendizaje automático:** Si los datos faltantes no son demasiado numerosos, puedes utilizar modelos de aprendizaje automático, como regresión lineal o KNN, para predecir los valores faltantes en función de otras variables.

**Imputación por agrupación (group imputation): **Para variables categóricas, puedes imputar los valores faltantes basándote en la agrupación de otras variables relacionadas.

**Imputación múltiple (multiple imputation):** Este enfoque implica generar múltiples conjuntos de datos imputados y combinar los resultados para reducir el sesgo. Es útil cuando los valores faltantes no son completamente aleatorios.

**Análisis de patrones de valores faltantes:** Puedes examinar si los valores faltantes siguen algún patrón discernible y diseñar estrategias de imputación en consecuencia. Por ejemplo, si los datos faltantes están relacionados con fechas o eventos específicos, puedes usar la interpolación temporal para imputar valores.

**Creación de indicadores de valores faltantes:** En lugar de imputar los valores faltantes, puedes agregar columnas binarias que indiquen si los valores originales estaban presentes o ausentes. Esto permite conservar la información sobre la ausencia de datos.

**Validación de los métodos de imputación:** Es importante validar los métodos de imputación utilizados para asegurarse de que no introduzcan sesgos o distorsiones en los datos.
